```python
# Make the `agentlab` package importable and load environment variables.
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd().parent / "src"))

from agentlab.config import get_chat_model
llm = get_chat_model()
print("Using model:", llm.model_name)
```

## Basic chat

```python
resp = llm.invoke("In one sentence, what is an AI agent?")
print(resp.content)
```

## A prompt + model + parser chain (LCEL)

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

prompt = ChatPromptTemplate.from_template("Explain {concept} to a {audience} in 2 sentences.")
chain = prompt | llm | StrOutputParser()
print(chain.invoke({"concept": "vector embeddings", "audience": "10-year-old"}))
```

## Structured output

```python
from pydantic import BaseModel, Field

class City(BaseModel):
    name: str = Field(description="City name")
    country: str = Field(description="Country")
    fun_fact: str = Field(description="A short fun fact")

structured = llm.with_structured_output(City)
structured.invoke("Tell me about Kyoto.")
```
